## 1. Importing the Numerical Library

## 2. Scalars: Individual Numbers

```python
# Scalars: single numerical values

x = 3.14      # floating-point number
n = 42        # integer

print("x:", x, "| type:", type(x))
print("n:", n, "| type:", type(n))
```

### Intuition

A scalar is a single number.

It represents one quantity — not a collection, not a structure, just one value.

For example:
- temperature
- a single measurement
- a single weight in a model

---

### Mathematical View

A scalar is an element of $\mathbb{R}$.

- $3.14 \in \mathbb{R}$
- $42 \in \mathbb{Z} \subset \mathbb{R}$

This means:
- integers (ℤ) are a subset of real numbers (ℝ)
- all values used in computation can be treated within ℝ

---

### Why This Matters

Different types of numbers behave differently in computation:

- integers are exact
- floating-point numbers are approximations

Most machine learning systems rely on floating-point numbers because:
- they allow continuous values
- they support smooth updates during learning

---

### AI Connection

In machine learning:

- a single weight is a scalar  
- a bias term is a scalar  
- a loss value is a scalar  

Even though models operate on large arrays, everything is ultimately built from individual scalar values.

---

### Conceptual Tension

If a scalar is just one number,  
why is it not enough to represent meaning?

This question leads to the next concept: vectors.

## 3. Arrays: From Numbers to Representations

```python
# Arrays: collections of numbers

import numpy as np

embedding = np.array([0.34, -0.12, 0.88, 0.01, -0.55])

print("Embedding:", embedding)
print("Shape:", embedding.shape)
print("Dtype:", embedding.dtype)
```

### Intuition

A single number captures one quantity.  
But meaning is not one-dimensional.

To represent something complex (like a word), we need multiple numbers working together.

An array is a structured collection of numbers.

You can think of it as a point with multiple coordinates.

---

### Mathematical View

An array is a vector in $\mathbb{R}^d$.

- The array `[0.34, -0.12, 0.88, 0.01, -0.55]` is a vector in $\mathbb{R}^5$
- Each component is a real number
- The dimension $d$ is the number of elements

Formally:

$$
\mathbf{x} = (x_1, x_2, ..., x_d) \in \mathbb{R}^d
$$

---

### Why This Matters

Real-world data is multi-dimensional.

A single value cannot encode:
- meaning
- context
- relationships

Arrays allow us to represent:
- images (pixels)
- signals (time series)
- language (embeddings)

---

### AI Connection

In modern AI systems:

- each word is mapped to a vector (embedding)
- embeddings live in high-dimensional space (e.g., $\mathbb{R}^{768}$ or $\mathbb{R}^{12288}$)
- similarity between words is measured by distance between vectors

Example:

- “king” and “queen” are close in vector space  
- “king” and “banana” are far apart  

This is how meaning becomes geometry.

---

### Visual Imagination

Imagine each word as a point floating in a high-dimensional space.

Words with similar meanings are located near each other.

---

### Conceptual Tension

If meaning is a point in space,  
what defines movement in that space?

This leads to operations.

## 4. Operations: Transforming Representations

```python
# Operations on arrays

scaled = embedding * 2
shifted = embedding + 1

print("Original:", embedding)
print("Scaled:", scaled)
print("Shifted:", shifted)
```

### Intuition

Operations change numbers.

When we apply an operation to an array, we transform all its components at once.

- scaling stretches the representation
- shifting moves it

Instead of changing one number, we change the entire structure.

---

### Mathematical View

Let $\mathbf{x} \in \mathbb{R}^d$.

Operations act component-wise:

- scaling:
  $$
  \mathbf{x}' = 2\mathbf{x}
  $$

- shifting:
  $$
  \mathbf{x}' = \mathbf{x} + 1
  $$

This means:
$$
x_i' = 2x_i \quad \text{or} \quad x_i' = x_i + 1
$$

for each component $i$.

These are simple transformations in $\mathbb{R}^d$.

---

### Why This Matters

Operations define how representations evolve.

Without operations:
- numbers remain static
- no transformation occurs
- no learning is possible

All computation in AI is built from repeated application of simple operations.

---

### AI Connection

In machine learning:

- neural networks apply operations layer by layer  
- each layer transforms the input representation  
- these transformations reshape data in high-dimensional space  

For example:
- scaling resembles weight application  
- shifting resembles bias addition  

Together, they form the core of linear transformations.

---

### Visual Imagination

Imagine a point in space.

Operations move this point:
- scaling stretches its distance from the origin  
- shifting moves it in a specific direction  

Learning is a sequence of such movements.

---

### Conceptual Tension

If operations move representations,  
what determines the direction and size of that movement?

This leads to gradients and learning rules.

## 5. Floating-Point Precision: Approximating Reality

```python
# Floating-point precision

e32 = embedding.astype(np.float32)
e16 = embedding.astype(np.float16)

print("Original dtype:", embedding.dtype)
print("float32 dtype:", e32.dtype)
print("float16 dtype:", e16.dtype)

print("\nValues (float32):", e32)
print("Values (float16):", e16)
```

### Intuition

Computers cannot store real numbers exactly.

Instead, they store approximations using a limited number of bits.

- more bits → more precision  
- fewer bits → less precision  

This means the same number can be represented with different levels of accuracy.

---

### Mathematical View

In theory, values lie in $\mathbb{R}$ (continuous and infinite precision).

In practice, computers use a finite subset of $\mathbb{R}$:

- float32 → 32-bit approximation  
- float16 → 16-bit approximation  

So we do not work with exact real numbers, but with discrete approximations of them.

---

### Why This Matters

Precision affects:

- numerical accuracy  
- memory usage  
- computational speed  

Lower precision:
- uses less memory  
- runs faster  
- but introduces rounding errors  

Higher precision:
- is more accurate  
- but more expensive  

---

### AI Connection

In modern AI systems:

- float32 is commonly used during training  
- float16 (or mixed precision) is used to:
  - reduce memory usage  
  - speed up computation  

Large models rely heavily on reduced precision to scale.

However, too little precision can:
- destabilize training  
- introduce numerical errors  

---

### Visual Imagination

Imagine trying to measure a length with a ruler:

- a precise ruler gives accurate measurements  
- a rough ruler gives approximations  

Floating-point precision is like choosing the resolution of your ruler.

---

### Conceptual Tension

If learning depends on precise updates,  
what happens when precision is reduced?

This leads to the mechanics of learning itself.

## 6. Learning Step: Updating a Weight

```python
# A simple weight update

w = np.float32(0.800)     # current weight
alpha = np.float32(0.010) # learning rate
g = np.float32(2.500)     # gradient

w_new = w - alpha * g

print("Old weight:", w)
print("New weight:", w_new)
```

### Intuition

A model learns by adjusting its parameters.

Each update changes a number slightly, moving it in a direction that reduces error.

This step is small, but it is repeated many times.

---

### Mathematical View

The update rule is:

$$
w_{\text{new}} = w_{\text{old}} - \alpha \cdot g
$$

where:

- $w$ is the parameter (a scalar in $\mathbb{R}$)  
- $\alpha$ is the learning rate  
- $g$ is the gradient  

This is a step of **gradient descent**.

The gradient tells us how the error changes with respect to the parameter.

---

### Why This Matters

Learning is not magic.

It is the repeated application of simple numerical updates.

A large model contains millions or billions of such parameters, all updated using rules like this one.

---

### AI Connection

In neural networks:

- every weight is updated using a rule like this  
- gradients are computed using backpropagation  
- updates are applied layer by layer  

This single equation is executed millions of times during training.

---

### Visual Imagination

Imagine standing on a landscape:

- the height represents error  
- the slope tells you which direction to move  

The update moves you slightly downhill.

---
