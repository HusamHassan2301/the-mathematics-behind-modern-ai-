This notebook shows how the same concepts from NumPy appear in real AI systems using PyTorch.

A tensor is PyTorch’s version of an array — but with additional capabilities for building and training models.

We will mirror NumPy:

*   Import PyTorch
*   Scalars (tensors)
*   Embedding layer
*   Lookup

## 1. Import PyTorch

```python
import torch
import torch.nn as nn
```

## 1. Importing the Deep Learning Framework

### Intuition

PyTorch is a framework used to build and train AI models.

It allows us to work with numbers in a structured and efficient way.

---

### Mathematical View

PyTorch works with tensors, which are elements of $\mathbb{R}^d$, $\mathbb{R}^{m \times n}$, and higher dimensions.

A tensor generalizes:
- scalars (ℝ)
- vectors (ℝ^d)
- matrices (ℝ^{m×n})

---

### AI Connection

PyTorch is used to implement:
- neural networks  
- embeddings  
- training algorithms  

It is the standard tool used in modern deep learning systems.

## 2. Scalars as Tensors

```python
x = torch.tensor(3.14)
n = torch.tensor(42)

print("x:", x, "| dtype:", x.dtype)
print("n:", n, "| dtype:", n.dtype)
```

## 2. Scalars as Tensors

### Intuition

A tensor can represent a single number.

Even one value is treated as a structured object.

---

### Mathematical View

A scalar tensor is an element of $\mathbb{R}$.

It is a zero-dimensional tensor.

---

### AI Connection

In neural networks:
- a weight can be a scalar  
- a bias can be a scalar  
- loss values are scalars  

PyTorch represents all of these as tensors for consistency.

## 3. Embedding Layer

```python
embed = nn.Embedding(num_embeddings=10, embedding_dim=4)

print("Embedding matrix shape:", embed.weight.shape)
print("Data type:", embed.weight.dtype)
```

## 3. Embedding Layer

### Intuition

An embedding layer is a table of numbers.

Each row represents a word (or token).

---

### Mathematical View

The embedding matrix is:

$$
W \in \mathbb{R}^{V \times d}
$$

where:
- $V$ = vocabulary size  
- $d$ = embedding dimension  

Each row is a vector in $\mathbb{R}^d$.

---

### AI Connection

This is the first layer in a language model.

- input: integer token ID  
- output: vector representation  

This is how words become numbers.

## 4. Lookup

```python
token_idx = torch.tensor([3])

vector = embed(token_idx)

print("Vector:", vector)
print("Shape:", vector.shape)
```

## 4. Embedding Lookup

### Intuition

We select a row from the embedding table.

That row becomes the representation of the word.

---

### Mathematical View

Given:

$$
W \in \mathbb{R}^{V \times d}
$$

Selecting index $i$ returns:

$$
\mathbf{x}_i \in \mathbb{R}^d
$$

---

### AI Connection

This is what happens in every language model:

1. word → integer  
2. integer → vector  

This single operation is the entry point to the entire model.

## 5. Operations

```python
scaled = vector * 2
shifted = vector + 1

print("Scaled:", scaled)
print("Shifted:", shifted)
```

## 5. Operations on Tensors

### Intuition

We transform the representation by applying operations.

---

### Mathematical View

If $\mathbf{x} \in \mathbb{R}^d$, then:

- scaling: $2\mathbf{x}$  
- shifting: $\mathbf{x} + 1$  

---

### AI Connection

Neural networks apply many such transformations layer by layer.

These operations reshape data in high-dimensional space.

## 6. Weight Update

```python
w = torch.tensor(0.800)
alpha = torch.tensor(0.010)
g = torch.tensor(2.500)

w_new = w - alpha * g

print("Old:", w.item())
print("New:", w_new.item())
```

## Learning Step

### Intuition

A model learns by adjusting its parameters.

---

### Mathematical View

$$
w_{\text{new}} = w - \alpha g
$$

---

### AI Connection

This update is applied to every parameter during training.

This is the mechanism behind learning.

## 7. Final Insight

PyTorch does not introduce new mathematics.

It implements the same ideas you saw in NumPy — but at scale.

The difference is not in the math.  
The difference is in automation, efficiency, and scale.

---

If numbers define position,  
operations define movement,  
then learning is controlled movement in $\mathbb{R}^d$.
